# For loops in Python

### ITERATION "Looping"

Two types of loops:

1.) Counting loops             FOR

Computers count from 0
Every is 0 indexed

0, 1, 2, 3, 4, 5...

h=1

e=2

l=3

l=4

o=5

!=6

"l" is in the 3rd place, but also place number 2

For loops give 
- the item
- the number of the item

Always runs x number of times

Never use counting loops if it can be stopped early


``` python
for counter in range(0, 10, 1)
```
for = python keyword
counter = name of variable
in = python keyword
range = built-in function   Needs three bits of information (parameters) START, STOP, STEP

Start: Default is 0
Stop: Up too but not included. (This has to be included)
Step: Default is 1







2.) Conditional loops         WHILE

Will stop on a condition(s)
Will start on a condition(s)

Stops as soon as a condition is met (or many conditions)




Infinite Loops (Never do this)
